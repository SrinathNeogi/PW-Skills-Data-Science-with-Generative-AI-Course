```python
#Q3
import matplotlib.pyplot as plt

plt.figure(figsize = (10,8))

categories = ['A', 'B', 'C', 'D', 'E']
values = [25, 40, 30, 35, 20]

plt.bar(categories,values , color = ['red' , 'blue' , 'green' , 'orange' , 'magenta'])

for i, v in enumerate(values):
    plt.text(i, v+1, str(v), ha='center')

plt.xlabel("Categories")
plt.ylabel("Values")

plt.title("Bar Chart")

plt.show()
```

# **Bokeh**

```python
#Q1
import bokeh.io
import bokeh.plotting
from bokeh.plotting import figure, output_file, show
import numpy as np

X = np.random.randint(1,10,10)

Y = np.sin(X)

output_file('test.html')

fig = figure(title = 'Test Bokeh Plot')

fig.scatter(X , Y)

show(fig)
```

```python
#Q2
x = np.random.randint(1,100,1000)
y = np.random.randint(1,100,1000)

fig = figure(title = 'Bokeh Plot')

output_file("test.html")

fig.scatter(x,y, size = 15, fill_color = 'red')

show(fig)
```

```python
#Q3
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

output_file('test.html')

fig = figure(title = "Bokeh bar plot" , x_range = fruits , toolbar_location = None , tools = "")

fig.vbar(x = fruits , top = counts , width = 0.5 , color = 'blue')

fig.xgrid.grid_line_color = None
fig.y_range.start = 0

fig.xaxis.axis_label = "Fruits"
fig.yaxis.axis_label = "Count"

show(fig)
```

```python
#Q4
data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)

output_file('test.html')

fig = figure(title = "Bokeh Histogram")

fig.quad(top = hist , bottom = 0 , left = edges[:-1] , right = edges[1:] , fill_color = 'blue' , line_color = 'white')

show(fig)
```

```python
#Q5
import numpy as np
from bokeh.io import output_file, show
from bokeh.plotting import figure
from bokeh.models import ColumnDataSource, ColorBar
from bokeh.transform import linear_cmap
from bokeh.palettes import Viridis256

# Create output file
output_file("bokeh_heatmap.html")

# Data for the heatmap
data_heatmap = np.random.rand(10, 10)  # 10x10 matrix of random values
x = np.linspace(0, 1, 10)              # x-coordinates
y = np.linspace(0, 1, 10)              # y-coordinates
xx, yy = np.meshgrid(x, y)

# Flatten the data to 1D arrays for Bokeh
x_coords = xx.flatten()
y_coords = yy.flatten()
heatmap_values = data_heatmap.flatten()

# Create a ColumnDataSource
source = ColumnDataSource(data=dict(
    x=x_coords,
    y=y_coords,
    value=heatmap_values
))

# Create a figure
fig = figure(
    title="Bokeh Heatmap",
    x_axis_label="X",
    y_axis_label="Y",
    x_range=(x.min(), x.max()),
    y_range=(y.min(), y.max()),
    toolbar_location=None,
    tools=""
)

# Add rectangles for the heatmap
mapper = linear_cmap(field_name="value", palette=Viridis256, low=heatmap_values.min(), high=heatmap_values.max())
fig.rect(
    x="x",
    y="y",
    width=0.1,
    height=0.1,
    source=source,
    fill_color=mapper,
    line_color=None
)

# Add a color bar to indicate value ranges
color_bar = ColorBar(color_mapper=mapper['transform'], width=8, location=(0, 0))
fig.add_layout(color_bar, 'right')

# Show the plot
show(fig)

```

# **Plotly**

```python
import numpy as np
import pandas as pd
import plotly.graph_objects as go
import plotly.express as px
```

```python
#Q1
np.random.seed(30)
data = {
    'X': np.random.uniform(-10, 10, 300),
    'Y': np.random.uniform(-10, 10, 300),
    'Z': np.random.uniform(-10, 10, 300)
}
df = pd.DataFrame(data)

fig = go.Figure()

fig.add_trace(go.Scatter3d(
    x = df['X'],
    y = df['Y'],
    z = df['Z'],
    mode = 'markers',
    marker = dict(
        size = 10,
        color = df['Z'],
        opacity = 0.8,
        colorscale = 'Viridis'
    )
))
fig.update_layout(
    scene = dict(
        xaxis_title = 'X',
        yaxis_title = 'Y',
        zaxis_title = 'Z'
    ),
    title = '3D Scatter Plot by Plotly'
)
fig.show()
```

```python
#Q2
np.random.seed(15)
data = {
    'Grade': np.random.choice(['A', 'B', 'C', 'D', 'F'], 200),
    'Score': np.random.randint(50, 100, 200)
}
df = pd.DataFrame(data)

import plotly.graph_objects as go

fig = go.Figure()

fig.add_trace(go.Violin(x = data['Grade'] , y = data['Score'] ,box_visible = True , meanline_visible = True , line_color = 'red'))

fig.update_layout(
    title = 'Violin Plot',
    xaxis_title = 'Grades',
    yaxis_title = 'Scores',
    showlegend = True
)
fig.show()
```

```python
#Q3
np.random.seed(20)
data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}
df = pd.DataFrame(data)
#print(df)
import plotly.express as px
pivot_table = df.pivot_table(
    index='Day',
    columns='Month',
    values='Sales',
    aggfunc='mean'
)
#print(pivot_table)
fig = px.imshow(
    pivot_table,
    labels=dict(x="Month", y="Day", color="Average Sales"),
    x=pivot_table.columns,  # Months
    y=pivot_table.index,    # Days
    title="Heatmap of Average Sales by Month and Day",
    color_continuous_scale='Viridis'
)

fig.show()
```

```python
#Q4
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))
data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}
df = pd.DataFrame(data)

fig = go.Figure()

fig.add_trace(go.Scatter3d(
    x = df['X'],
    y = df['Y'],
    z = df['Z'],

    mode = 'markers',

    marker = dict(
        size = 10,
        color = df['Z'],
        opacity = 0.8,
        colorscale = 'Viridis'
    )
))
fig.update_layout(
    title = "3D Scatter plot",

    scene = dict(
        xaxis_title = 'X',
        yaxis_title = 'Y',
        zaxis_title = 'Z'
    )
)
fig.show()
```
