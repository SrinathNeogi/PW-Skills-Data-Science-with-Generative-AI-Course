```python
# 4 -> Implement a program to simulate the rolling of a fair six-sided die and calculate the expected value and variance of the outcomes
import numpy as np

# Define six-sided dice values and equal probabilities
dice_values = np.array([1, 2, 3, 4, 5, 6])
probabilities = np.full(6, 1/6)  # All probabilities are 1/6

# Display values and probabilities
for v, p in zip(dice_values, probabilities):
    print(f"{v} : {p:.4f}")

# Calculate Expected Value (Mean)
expected_value = np.sum(dice_values * probabilities)
print(f"Expected Value (Mean) = {expected_value}")

# Calculate Variance
variance = np.sum(probabilities * (dice_values - expected_value) ** 2)
print(f"Variance = {variance:.2f}")
```

```python
# 5 -> Create a Python function to generate random samples from a given probability distribution (e.g., binomial, Poisson) and calculate their mean and variance.
def generate_samples(type , features, size):
    if type == 'binomial':
        """features[0] = trial
        feature[1] = success rate"""
        sample = np.random.binomial(features[0], features[1], size)

    elif type == 'poisson':
        """feature[0] = rate - > average number of events per interval"""
        sample = np.random.poisson(features[0] , size)

    #print(sample)
    return np.mean(sample),np.var(sample)

print(f"Binomial sample mean and varianc = {generate_samples('binomial' , [1000,0.5], 1000)}")
print(f"Poisson sample mean and varianc = {generate_samples('poisson' , [100], 1000)}")
```

```python
# 11 A company wants to test if a new website layout leads to a higher conversion rate (percentage of visitors who make a purchase). They collect data from the old and new layouts to compare.
import numpy as np

# 50 purchases out of 1000 visitors

old_layout = np.array([1] * 50 + [0] * 950)

# 70 purchases out of 1000 visitors

new_layout = np.array([1] * 70 + [0] * 930)
# Null Hypothesis -> Conversion is not effected.
# Alternate Hypothesis -> Conversion is increased.
from statsmodels.stats.weightstats import ztest

zscore , pvalue = ztest(
    new_layout,
    old_layout,
    alternative = 'larger'
)
print(f"P-value = {pvalue}")
if(pvalue < 0.05):
    print("Reject the Null Hypothesis -> There is enough evidence that the new layout leads to higher conversion.")
else:
    print("Fail to Reject Null Hypothesis -> Conversion rate is not increased.")
```

```python
# 12 A tutoring service claims that its program improves students' exam scores. A sample of students who participated in the program was taken, and their scores before and after the program were recorded.

before_program = np.array([75, 80, 85, 70, 90, 78, 92, 88, 82, 87])

after_program = np.array([80, 85, 90, 80, 92, 80, 95, 90, 85, 88])

"""
In this case we have to compare the two samples before and after
We will take the difference of the two samples
Null Hypothesis => There is no difference between the marks of students => difference = 0 => H0 = 0
Alternate Hypothesis => The difference of the marks is not 0 => H1 > 0
"""
from statsmodels.stats.weightstats import ztest
difference = after_program - before_program

zscore , pvalue = ztest(
    difference, # Variable (sample) in which the hypothesis should be tested
    value = 0, # The Null hypothesis is difference = 0
    alternative = 'larger'
)
print(f"P - value = {pvalue}")

if pvalue < 0.05:
    print("Reject Null Hypothesis -> There is significant evidences that the tuition program increased the students' marks.")
else:
    print("There is no difference between marks before and after the tuition program.")
```

```python
# 15 A company is testing two different website layouts to see which one leads to higher click-through rates. Write a Python function to perform an A/B test analysis, including calculating the t-statistic, degrees of freedom, and p-value.

layout_a_clicks = [28, 32, 33, 29, 31, 34, 30, 35, 36, 37]

layout_b_clicks = [40, 41, 38, 42, 39, 44, 43, 41, 45, 47]
"""
Null Hypothesis => There is no significant difference between the click-through rate. -> mean1 = mean2
Alternate Hypothesis => There is a significant difference between the two website layouts.
"""
import scipy.stats as stats

def ab_t_test(sample1 , sample2):
    tstats , pvalue = stats.ttest_ind(sample1 , sample2, equal_var = True)
    print(f"T-statistic = {tstats}")
    print(f"P-value = {pvalue}")

    print(f"Degrees of freedom = {len(sample1)-1} and {len(sample2)-1}")

    alpha = 0.05

    if pvalue <= 0.05:
        print("Reject Null Hypothesis -> There is a significant difference between the two website layouts.")
    else:
        print("Fail to reject Null Hypothesis -> There is no significant difference between the two website layouts.")

ab_t_test(layout_a_clicks , layout_b_clicks)
```

```python
# 21 A political analyst wants to determine if there is a significant association between age groups and voter preferences (Candidate A or Candidate B). They collect data from a sample of 500 voters and classify them into different age groups and candidate preferences. Perform a Chi-Square test to determine if there is a significant association between age groups and voter preferences.

import numpy as np
from scipy.stats import chi2_contingency
import pandas as pd

# Set seed for reproducibility
np.random.seed(0)

# Generate synthetic data (corrected to 500 voters)
age_groups = np.random.choice(["18-30", "31-50", "51+"], size=500)
voter_preferences = np.random.choice(["Candidate A", "Candidate B"], size=500)

# Hypotheses
"""
Null Hypothesis (H0): There is no significant association between age groups and voter preferences.
Alternative Hypothesis (H1): There is a significant association between age groups and voter preferences.
"""

# Create observed frequency table (contingency table)
observed_table = pd.crosstab(age_groups, voter_preferences)
print("Observed Frequencies Table:\n", observed_table)

# Perform Chi-Square test
chi2_stat, pvalue, dof, expected_table = chi2_contingency(observed_table)

# Convert expected frequencies into a DataFrame for better readability
expected_df = pd.DataFrame(expected_table, index=observed_table.index, columns=observed_table.columns)
print("\nExpected Frequencies Table:\n", expected_df)

# Print test results
print(f"\nChi-Square Statistic: {chi2_stat:.4f}")
print(f"Degrees of Freedom: {dof}")
print(f"P-value: {pvalue:.4f}")

# Conclusion
alpha = 0.05
if pvalue <= alpha:
    print("Reject Null Hypothesis => There is strong evidence that age groups and voter preferences are associated.")
else:
    print("Fail to reject Null Hypothesis => No significant association between age groups and voter preferences.")
```
